### Naive Bayes Classifier

Naive bayes classifier works on the pronciples of conditional probability as given by the bayes' theorem. \
Bayes' theorem basically calculates the conditional probability of the occurrence of an event based on prior knowledge of conditions that might be related to the event. \
\
Naive Bayes Classifier > Classification > Supervised learning

### Naive Bayes Classifier used for
- Face recognition 
- Weather prediction
- Medical Diagnosis
- News Classification

### Advantages of Naive Bayes Classifier
1. Very simple and easy to implement
2. Needs less training data
3. Handles both continuous and discrete data
4. Highly scalable with number of predictors and data points
5. As it is fast, it can be used in real time predictions
6. Not sensitive to irrelevant features

```python
# Loading the necessary 
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

from sklearn.datasets import fetch_20newsgroups

data = fetch_20newsgroups()
data.target_names
```

```python
#Defining all the categories
categories = ['alt.atheism', 'comp.graphics', 'comp.os.ms-windows.misc',
              'comp.sys.ibm.pc.hardware', 'comp.sys.mac.hardware',
              'comp.windows.x', 'misc.forsale', 'rec.autos',
              'rec.motorcycles', 'rec.sport.baseball', 
              'rec.sport.hockey', 'sci.crypt', 'sci.electronics',
              'sci.med', 'sci.space', 'soc.religion.christian',
              'talk.politics.guns', 'talk.politics.mideast',
              'talk.politics.misc', 'talk.religion.misc']

#Training the data on these categories
train = fetch_20newsgroups(subset = 'train', categories = categories)
```

```python
#Testing the data for these categories
test = fetch_20newsgroups(subset = 'test', categories = categories)
```

```python
#Importing necessary packages
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline

#Creating a model based Multinomial Naive Bayes
model = make_pipeline(TfidfVectorizer(), MultinomialNB())
```

```python
#Training the model with the train data
model.fit(train.data, train.target)

#Creating labels for the test data
labels = model.predict(test.data)
```

```python
#Creating confusion matrix and heat map
from sklearn.metrics import confusion_matrix

matriks = confusion_matrix(test.target, labels)
hm      = sns.heatmap(matriks.T, square = True, annot = True, 
            fmt = 'd', cbar = False, 
            xticklabels = train.target_names,
            yticklabels = train.target_names)

axes = hm.axes
axes.set_ylim(matriks.T.shape[0], 0)

#Plotting Heatmap of Confusion Matrix
plt.xlabel('true_label')
plt.ylabel('predicted_label')
```

```python
#Predicting category on new data based on trained model
def predict_category(s, train = train, model = model) :
    pred = model.predict([s])
    return train.target_names[pred[0]]
```

```python
predict_category('Michael Jordan')
```

```python
predict_category('Ferrari is fast')
```
